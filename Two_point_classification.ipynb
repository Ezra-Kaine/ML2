```python
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
df=pd.read_csv('insurance_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.age,df.bought_insurance,marker='*',color='red')
```

```python
from sklearn.model_selection import  train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['age']],df.bought_insurance,test_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predicted=model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test,y_test)
```

```python
import math
def signoid(x):
  return 1/(1+math.exp(-x))
```

```python
def prediction_function(age):
  coef=0.36494784
  intercept=-15.441047
  z=coef*age+intercept
  y_predicted=signoid(z)
  return y_predicted
```

```python
def main():
  age=int(input("Enter the age:"))
  y_predicted=prediction_function(age)
  if y_predicted>0.5:
    print("The person will buy insurance")
  else:
    print("The person will not buy insurance")
```

```python
main()
```
